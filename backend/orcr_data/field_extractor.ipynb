```python
import pandas as pd
```

```python
df=pd.read_csv('ORCR_16_22_all.csv')
df
```

```python
institutes=df["Institute"].unique()
branches=df["Academic Program Name"].unique()
quota=df["Quota"].unique()
seat=df["Seat Type"].unique()
gender=df["Gender"].unique()
orank=df["Opening Rank"].unique()
crank=df["Closing Rank"].unique()
year=df["Year"].unique()
round=df["Round"].unique()
```

```python
with open('allbranch.txt','w') as file:
    file.write('\n'.join(branches))
with open('allinstitutes.txt','w') as file:
    file.write('\n'.join(institutes))

with open('allseat.txt','w') as file:
    file.write('\n'.join(seat))
# with open('allor.txt','w') as file:
#     file.write('\n'.join(orank))
# with open('allcr.txt','w') as file:
#     file.write('\n'.join(crank))

with open('allquota.txt','w') as file:
    for qt in quota:
        file.write(str(qt)+'\n')
with open('allyear.txt','w') as file:
    for yr in year:
        file.write(str(yr)+'\n')

with open('allround.txt','w') as file:
    for rnd in round:
        file.write(str(rnd)+'\n')

with open('allgender.txt','w') as file:
    for gen in gender:
        file.write(str(gen)+'\n')

```
